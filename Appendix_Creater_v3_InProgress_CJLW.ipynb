```python
# Write the code to merge Documents here
from PyPDF2 import PdfFileReader, PdfFileWriter

# Dont move your files after step 1!! This step will only work if you keep your sorted files in the same location!

Main_dir = os.listdir(dirname)


# Merge Appendix Daily Reports
def merge_pdfs(paths, output):
    pdf_writer = PdfFileWriter()

    for path in paths:
        pdf_reader = PdfFileReader(path)
        for page in range(pdf_reader.getNumPages()):
            # Add each page to the writer object
            pdf_writer.addPage(pdf_reader.getPage(page))

    # Write out the merged PDF
    with open(output, 'wb') as out:
        pdf_writer.write(out)

if __name__ == '__main__':
    paths = Directory_DailyReport
    merge_pdfs(paths, output='Complete_DailyReports.pdf')

# Merge Appendix Daily Tailgate
def merge_pdfs(paths, output):
    pdf_writer = PdfFileWriter()

    for path in paths:
        pdf_reader = PdfFileReader(path)
        for page in range(pdf_reader.getNumPages()):
            # Add each page to the writer object
            pdf_writer.addPage(pdf_reader.getPage(page))

    # Write out the merged PDF
    with open(output, 'wb') as out:
        pdf_writer.write(out)

if __name__ == '__main__':
    paths = Directory_Tailgate
    merge_pdfs(paths, output='Complete_DailyTailgates.pdf')
    

# Merge Appendix Daily JHA
def merge_pdfs(paths, output):
    pdf_writer = PdfFileWriter()

    for path in paths:
        pdf_reader = PdfFileReader(path)
        for page in range(pdf_reader.getNumPages()):
            # Add each page to the writer object
            pdf_writer.addPage(pdf_reader.getPage(page))

    # Write out the merged PDF
    with open(output, 'wb') as out:
        pdf_writer.write(out)

if __name__ == '__main__':
    paths = Directory_JHA
    merge_pdfs(paths, output='Complete_DailyJHA.pdf')

# Add a Merge for Equipment Certs


# Don't move the files into a seperate folders, otherwise the next step will not work.  

"""

Important!!!!!!!

Once this is done, you cant run the code again without getting a file already exists ERROR Raised, doesnt matter if you move the file out of the folder. You would need to delete the file or RENAME your output file


"""

    





```
